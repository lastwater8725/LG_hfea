```python
train_df = train
test_df = test
```

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler
```

```python
# ===================
# 2. 결측치 탐색 및 제거
# ===================
thresh = 0.8  # 80% 이상 결측치 컬럼 제거
missing_train = train_df.isnull().sum() / len(train_df)
drop_cols = missing_train[missing_train > thresh].index.to_list()
train_df.drop(columns=drop_cols + ["ID"], inplace=True, errors='ignore')
test_df.drop(columns=[col for col in drop_cols if col in test_df.columns] + ["ID"], inplace=True, errors='ignore')

# ===================
# 3. 범주형 변수 처리
# ===================
categorical_cols = train_df.select_dtypes(include=['object']).columns

# 특정 시술 유형 그룹화
rare_categories = train_df["특정 시술 유형"].value_counts()[train_df["특정 시술 유형"].value_counts() < 500].index
train_df["특정 시술 유형"] = train_df["특정 시술 유형"].replace(rare_categories, "기타")
test_df["특정 시술 유형"] = test_df["특정 시술 유형"].replace(rare_categories, "기타")

# Label Encoding 적용 (Train + Test 데이터 결합 후 fit)
combined_df = pd.concat([train_df, test_df])
label_encoders = {}
for col in ["시술 시기 코드", "시술 유형", "특정 시술 유형", "배란 유도 유형", "난자 출처", "정자 출처"]:
    le = LabelEncoder()
    combined_df[col].fillna("Unknown", inplace=True)  # 결측치 처리
    combined_df[col] = le.fit_transform(combined_df[col])
    train_df[col] = le.transform(train_df[col])
    test_df[col] = le.transform(test_df[col])
    label_encoders[col] = le

# ===================
# 4. 이상치 처리 (로그 변환 적용)
# ===================
numeric_cols = train_df.select_dtypes(include=['number']).columns
for col in numeric_cols:
    if train_df[col].skew() > 2:
        train_df[col] = np.log1p(train_df[col])
        if col in test_df.columns:
            test_df[col] = np.log1p(test_df[col])

# ===================
# 5. 결측치 처리
# ===================
for col in train_df.select_dtypes(include=['number']).columns:
    median_value = train_df[col].median()
    train_df[col].fillna(median_value, inplace=True)
    if col in test_df.columns:
        test_df[col].fillna(median_value, inplace=True)
```
